Sentimental Analysis


```python
import pandas as pd
```

```python
imdb = pd.read_csv('/content/IMDB Dataset.csv')
```

```python
imdb.info()
```

```python
imdb.head()
```

```python
imdb.isnull().sum()
```

```python
a=imdb['review']
b=imdb['sentiment']
```

```python
from sklearn.model_selection import train_test_split
atrain,atest,btrain,btest=train_test_split(a,b,test_size=0.3,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
conv=CountVectorizer(stop_words='english')
```

```python
atrain_V = conv.fit_transform(atrain)
atest_v = conv.transform(atest)
```

For Text Classification, we will use Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
MNB = MultinomialNB()
MNB
```

```python
MNB.fit(atrain_V, btrain)
```

```python
b_pred = MNB.predict(atest_v)
```

```python
b_pred
```

```python
btest
```

```python
from sklearn.metrics import accuracy_score
```

```python
ac = accuracy_score(b_pred,btest)
ac
```

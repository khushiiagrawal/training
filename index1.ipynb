```python
pred = LR.predict(xtest)
```

```python
pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(pred, ytest)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('/content/student_exam_data.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
plt.xlabel('Study Hours')
plt.ylabel('Pass/Fail')
plt.title('Student Dataset')
plt.plot(df['Study Hours'],df['Pass/Fail'],'b*')
```

```python
A=df.drop('Pass/Fail',axis=1) #axis=1-dele col, axis=0-del row
B=df['Pass/Fail']
```

```python
Atrain,Atest,Btrain,Btest=train_test_split(A,B,test_size=0.3, random_state=4)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LGR=LogisticRegression()
LGR
```

```python
LGR.fit(Atrain, Btrain)
```

```python
ypred=LGR.predict(Atest)
ypred
```

```python
Btest
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=accuracy_score(ypred, Btest)
acc
```
